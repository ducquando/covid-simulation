```python
from SIRmodel import SIR
```

```python
# Set params
networks = ["random", "scale-free", "small-world"]
TIME = 30 # days
RATE_IMMUNE = 0.53
RATE_SI = 1 - RATE_IMMUNE
RATE_RI = 0.178
```

```python
# Scale-free network
sf = SIR(networks[1], time = TIME, rate_si = RATE_SI, rate_ir = RATE_RI)
sf.make_video("simulation.mp4")
```

```python
# Small world network
sw = SIR(networks[2], time = TIME, rate_si = RATE_SI, rate_ir = RATE_RI)
sw.make_video("simulation.mp4")
```

```python
# Random network
rd = SIR(networks[0], time = TIME, rate_si = RATE_SI, rate_ir = RATE_RI)
rd.make_video("simulation.mp4")
```
